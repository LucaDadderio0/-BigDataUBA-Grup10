```python
#Resolucion del ejercicio 2
#Crear variable educ
def calcular_educ(row):
    nivel = row["CH12"]
    completo = row["CH13"]
ultimo_anio = row["CH14"]
    
    if pd.isna(nivel) or pd.isna(completo) or pd.isna(ultimo_anio):
        return None

    if nivel == 1:  # Jardín
        return 0
    if nivel == 2:  # Primaria
        return 7 if completo == 1 else ultimo_anio
    if nivel == 3:  # EGB
        return 9 if completo == 1 else ultimo_anio
    if nivel in [4, 5]:  # Secundaria / Polimodal
        return 12 if completo == 1 else 7 + ultimo_anio
    if nivel == 6:  # Terciario
        return 15 if completo == 1 else 12 + ultimo_anio
    if nivel == 7:  # Universitario
        return 17 if completo == 1 else 13 + ultimo_anio
    if nivel == 8:  # Posgrado
        return 20
return None
# Asegurarse que CH13 y CH14 sean numéricos
df_unificado["CH13"] = pd.to_numeric(df_unificado["CH13"], errors="coerce")
df_unificado["CH14"] = pd.to_numeric(df_unificado["CH14"], errors="coerce")
df_unificado["CH12"] = pd.to_numeric(df_unificado["CH12"], errors="coerce")
```
