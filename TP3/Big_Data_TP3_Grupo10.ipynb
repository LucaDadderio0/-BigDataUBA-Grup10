```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
# Leer los archivos
df_2024 = pd.read_excel("usu_individual_T124.xlsx")
df_2004 = pd.read_stata("Individual_t104.dta")
#Convertir los nombres de columnas a mayusculas
df_2024.columns = df_2024.columns.str.upper()
df_2004.columns = df_2004.columns.str.upper()
# Filtrar solo región 40
df_2024 = df_2024[df_2024["REGION"] == 40]
df_2004 = df_2004[df_2004["REGION"] == "NOA"]
#Homogeneizar los valores
df_2004["REGION"] = 40
df_2004['TRIMESTRE'] = df_2004['TRIMESTRE'].map({'1er. Trimestre': 1})
df_2004['CH03'] = df_2004['CH03'].map({'Jefe': 1 , 'Cónyuge/Pareja': 2, 'Hijo/Hijastro': 3, 'Yerno/Nuera': 4, 'Nieto': 5, 'Madre/Padre': 6, 'Suegro': 7, 'Hermano': 8, 'Otros familiares': 9, 'No familiares': 10}) 
df_2004['CH04'] = df_2004['CH04'].map({'Varón': 1, 'Mujer': 2})
df_2024['CH06'] = df_2024['CH06'].replace(0, np.nan)
df_2004['CH06'] = df_2004['CH06'].replace(0, np.nan)
df_2004['CH07'] = df_2004['CH07'].map({'Unido': 1, 'Casado': 2, 'Separado o divorciado': 3, 'Viudo': 4, 'Soltero': 5})
df_2024['CH12'] = df_2024['CH12'].replace(0, np.nan)
df_2004['CH12'] = df_2004['CH12'].map({'Jardin/Preescolar': 1 , 'Primario': 2, 'EGB': 3, 'Secundario': 4, 'Polimodal': 5, 'Terciario': 6, 'Universitario': 7, 'Posgrado Univ': 8, 'Educacion especial (discapacitado)': 9})                                               
df_2004['CH12'] = df_2004['CH12'].replace(0, np.nan)
df_2024['CH13'] = df_2024['CH13'].replace(0, np.nan)                                                                             
df_2004['CH13'] = df_2004['CH13'].map({'Si': 1, 'No': 2, 'Ns./Nr.': 9})
df_2004['CH13'] = df_2004['CH13'].replace(0, np.nan)
df_2004['CH14'].map({'Ninguno': 0 , 'Primario': 1, 'Segundo': 2 , 'Tercero': 3 , 'Cuarto': 4  , 'Quinto': 5  , 'Sexto': 6 , 'Septimo': 7  , 'Octavo': 8 , 'Noveno': 9 })
df_2004['NIVEL_ED'] = df_2004['NIVEL_ED'].map({'Primaria Incompleta (incluye educación especial)': 1, 'Primaria Completa': 2, 'Secundaria Incompleta': 3, 'Secundaria Completa': 4, 'Superior Universitaria Incompleta': 5, 'Superior Universitaria Completa': 6, 'Sin instrucción': 7, 'Ns./ Nr.': 9,})
df_2004['ESTADO'] = df_2004['ESTADO'].map({'Entrevista individual no realizada (no respuesta al cuestionario individual)': 0, 'Ocupado': 1, 'Desocupado': 2, 'Inactivo': 3, 'Menor de 10 años': 4})
df_2004['P21'].map({ 'Sin ingresos': 00})
df_2004['PP3E_TOT'].map({ 'Sin Horas': 0.0})
df_2004['PP3F_TOT'].map({ 'Sin Horas Extras': 00})
#Lista de las posibles variables a usar
variables_a_usar= ['CODUSU', 'NRO_HOGAR', 'COMPONENTE', 'ANO4', 'TRIMESTRE', 'REGION', 'CH03', 'CH04', 'CH06', 'CH07', 'CH12', 'CH13', 'CH14', 'ESTADO', 'P21', 'PP3E_TOT', 'PP3F_TOT', 'NIVEL_ED']
# Seleccionar las columnas de las variables a usar
df_2024 = df_2024[variables_a_usar]
df_2004 = df_2004[variables_a_usar]
display(df_2004)
display(df_2024)
```

```python
#Resolucion del Parte I
#Resolucion del ejercicio 1
#instalar pandas si no lo tienen. Debes sacar # de #!pip install pandas.
#!pip install pandas
#instalar matplotlib si no lo tienen. Debes sacar # de #!pip install matplotlib.
#!pip install matplotlib
#instalar seaborn si no lo tienen. Debes sacar # de #!pip install seaborn.
#!pip install seaborn
#importar pandas, matplotlib, seaborn y numpy
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
# Unir ambos archivos
df = pd.concat([df_2024, df_2004])

# Asegurar que CH06 sea numérico
df["CH06"] = pd.to_numeric(df["CH06"], errors="coerce")

# Crear la variable edad2
df["edad2"] = df["CH06"] ** 2

# Filtrar datos válidos: CH06 no nulo y ESTADO en 1 (Ocupado) o 2 (Desocupado)
df_plot = df[df["CH06"].notna() & df["ESTADO"].isin([1, 2])].copy()

# Mapear etiquetas legibles para graficar
estado_labels = {1: "Ocupado", 2: "Desocupado"}
df_plot["ESTADO_LABEL"] = df_plot["ESTADO"].map(estado_labels)

# Panel A: Histograma de edad
sns.histplot(df_plot["CH06"], bins=30, kde=False, color='skyblue')
plt.title("Panel A: Histograma de Edad")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()

# Panel B: Kernel por estado
sns.kdeplot(data=df_plot[df_plot["ESTADO_LABEL"] == "Ocupado"], x="CH06", label="Ocupado")
sns.kdeplot(data=df_plot[df_plot["ESTADO_LABEL"] == "Desocupado"], x="CH06", label="Desocupado")
plt.title("Panel B: Kernel de Edad por Estado")
plt.xlabel("Edad")
plt.legend()
plt.show()
```

```python
#Resolucion del ejercicio 2
# Asegurar tipo numérico
for col in ["CH12", "CH13", "CH14"]:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Calcular años de educación (educ)
def calcular_educ(row):
    n, c, a = row["CH12"], row["CH13"], row["CH14"]
    if pd.isna(n) or pd.isna(c) or pd.isna(a): return None
    if n == 1: return 0
    if n == 2: return 7 if c == 1 else a
    if n == 3: return 9 if c == 1 else a
    if n in [4, 5]: return 12 if c == 1 else 7 + a
    if n == 6: return 15 if c == 1 else 12 + a
    if n == 7: return 17 if c == 1 else 13 + a
    if n == 8: return 20
    return None

df["educ"] = df.apply(calcular_educ, axis=1)

# Estadísticas descriptivas
print(df["educ"].describe(percentiles=[0.5]))
```

```python
#Resolucion del ejercicio 3.a
# Ajuste por inflación para 2004
df["P21_adj"] = df.apply(
    lambda row: row["P21"] * 15 if row["ANO4"] == 2004 and pd.notna(row["P21"]) else row["P21"],
    axis=1
)

# Crear salario semanal
df["salario_semanal"] = df["P21_adj"] / 40

# Filtrar para gráfico (estados 1: Ocupado, 2: Desocupado)
df_sal = df[df["salario_semanal"].notna() & df["ESTADO"].isin([1, 2])]

# Recortar el 1% superior para evitar valores extremos
q99 = df_sal["salario_semanal"].quantile(0.99)
df_sal = df_sal[df_sal["salario_semanal"] < q99]

# Agregar etiquetas legibles
df_sal["ESTADO_LABEL"] = df_sal["ESTADO"].replace({1: "Ocupado", 2: "Desocupado"})

# Histograma
sns.histplot(df_sal["salario_semanal"], bins=40)
plt.title("Histograma de salario semanal")
plt.xlabel("Salario semanal (ajustado)")
plt.ylabel("Frecuencia")
plt.show()

# Filtrar grupos con suficiente variabilidad para KDE
df_sal_kde = df_sal.groupby("ESTADO_LABEL").filter(lambda x: x["salario_semanal"].nunique() > 1)

# KDE por estado laboral (solo si hay más de un valor distinto por grupo)
if not df_sal_kde.empty:
    sns.kdeplot(data=df_sal_kde, x="salario_semanal", hue="ESTADO_LABEL", common_norm=False)
    plt.title("Densidad del salario semanal por estado laboral")
    plt.xlabel("Salario semanal (ajustado)")
    plt.ylabel("Densidad")
    plt.show()
```

```python
#Resolucion del ejercicio 5
# Filtrar el DataFrame para el año indicado (2004 o 2024)
def resumen(df, año):
    d = df[df["ANO4"] == año]
    return {
        "Año": año,
        "Total": len(d),
        "Ocupados": (d["ESTADO"] == "Ocupado").sum(),
        "Desocupados": (d["ESTADO"] == "Desocupado").sum(),
        "edad2": d["edad2"].notna().sum(),
        "educ": d["educ"].notna().sum(),
        "salario_semanal": d["salario_semanal"].notna().sum(),
        "horastrab": d["horastrab"].notna().sum()
    }

tabla_resumen = pd.DataFrame([resumen(df, 2004), resumen(df, 2024)])
print(tabla_resumen)
```

```python
#Resolucion del Parte II
#Resolucion del ejercicio 1
#Definir las 5 variables que se usa en el PCA.
vars_pca = ["CH06", "edad2", "educ", "salario_semanal", "horastrab"]
# Filtrar las filas con datos completos
df_pca = df[vars_pca].dropna()
#Mostrar la matriz de correlaciones
print(df_pca.corr())
```

```python
#Resolucion del ejercicio 2
#Instalamos los paquetes necesarios
!pip install statsmodels
!pip install scikit-learn
!pip install ISLP
#importamos paquetes necesarios, excepto pandas, numpy y matplotlib ya que ya lo habiamos importado anteriormente.
import ISLP
from ISLP import load_data
from statsmodels.datasets import get_rdataset
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
#Estandarizar las variables
X = StandardScaler().fit_transform(df_pca)
#Aplicar PCA
pca = PCA()
scores = pca.fit_transform(X)

# Graficar PC1 vs PC2
plt.scatter(scores[:,0], scores[:,1], alpha=0.3)
plt.title("PCA - PC1 vs PC2")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
print("")
```

```python
#Resolucion del ejercicio 3
# Obtener cargas (loadings) de los dos primeros componentes
loadings = pd.DataFrame(
    pca.components_.T,
    index=["CH06", "edad2", "educ", "salario_semanal", "horastrab"],
    columns=[f"PC{i+1}" for i in range(len(pca.components_))]
)

# Mostrar las cargas de PC1 y PC2
print(loadings[["PC1", "PC2"]])
```

```python
#Resolucion del ejercicio 4
# Scree plot: Varianza explicada por cada componente
plt.plot(range(1, 6), pca.explained_variance_ratio_, marker='o')
plt.title("Scree Plot - Varianza explicada por componente")
plt.xlabel("Componente Principal")
plt.ylabel("Proporción de Varianza")
plt.show()
```

```python
#Resolucion del ejercicio 5.a
#importar los librerias necesarios, excepto pandas, numpy y matplotlib que ya lo habiamos importado anteriormente
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.cluster import KMeans

# Copiar las variables seleccionadas (CH06, edad2, educ, salario_semanal, horastrab) sin nulos.
df_kmeans = df_pca.copy()
#Estandarizar
X_std = StandardScaler().fit_transform(df_kmeans)

# Aplicar K-Means para distintos valores de k
for k in [2, 4, 10]:
    model = KMeans(n_clusters=k, n_init=20, random_state=0)
    df_kmeans[f"cluster_k{k}"] = model.fit_predict(X_std)

# Gráfico con k=2
sns.scatterplot(x=scores[:, 0], y=scores[:, 1], hue=df_kmeans["cluster_k2"], palette="Set2")
plt.title("KMeans (k=2) sobre PC1 y PC2")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
