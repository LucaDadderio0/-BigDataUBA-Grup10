```python
#Resolucion del ejercicio 2
# Asegurar tipo numérico
for col in ["CH12", "CH13", "CH14"]:
    df_2004[col] = pd.to_numeric(df_2004[col], errors="coerce")
    df_2024[col] = pd.to_numeric(df_2024[col], errors="coerce")

# Calcular años de educación (educ)
def calcular_educ(row):
    n, c, a = row["CH12"], row["CH13"], row["CH14"]
    if pd.isna(n) or pd.isna(c) or pd.isna(a): return None
    if n == 1: return 0
    if n == 2: return 7 if c == 1 else a
    if n == 3: return 9 if c == 1 else a
    if n in [4, 5]: return 12 if c == 1 else 7 + a
    if n == 6: return 15 if c == 1 else 12 + a
    if n == 7: return 17 if c == 1 else 13 + a
    if n == 8: return 20
    return None

df_2004["educ"] = df_2004.apply(calcular_educ, axis=1)
df_2024["educ"] = df_2024.apply(calcular_educ, axis=1)

# Estadísticas descriptivas
# Estadísticas descriptivas con tabla
print("\nEstadísticas descriptivas del 2004:")
print(df_2004["educ"].describe(percentiles=[0.5]))
# Estadísticas descriptivas con grafico
sns.histplot(df_2004["educ"], bins=20, kde=True, color="cornflowerblue")
plt.title("Distribución de años de educación - 2004")
plt.xlabel("Años de educación")
plt.ylabel("Frecuencia / Densidad")
plt.show()
# Estadísticas descriptivas con tabla
print("\nEstadísticas descriptivas del 2024:")
print(df_2024["educ"].describe(percentiles=[0.5]))
# Estadísticas descriptivas con grafico
sns.histplot(df_2024["educ"], bins=20, kde=True, color="mediumseagreen")
plt.title("Distribución de años de educación - 2024")
plt.xlabel("Años de educación")
plt.ylabel("Frecuencia / Densidad")
plt.show()
```

```python
#Resolucion del ejercicio 5
def tabla_resumen_custom(df_2004, df_2024):
    # Variables consideradas "limpias y homogeneizadas"
    vars_clave = ["CH06", "edad2", "educ", "salario_semanal", "horastrab", "ESTADO"]

    df_total = pd.concat([df_2004, df_2024], ignore_index=True)

    # Función auxiliar para contar filas sin nulos en todas las variables clave
    def contar_limpios(df):
        return df[vars_clave].dropna().shape[0]

    # Crear diccionario con filas descriptivas
    datos = {
        "Cantidad de observaciones": [
            len(df_2004),
            len(df_2024),
            len(df_total)
        ],
        "Cantidad de observaciones con NaN en ESTADO": [
            df_2004["ESTADO"].isna().sum(),
            df_2024["ESTADO"].isna().sum(),
            df_total["ESTADO"].isna().sum()
        ],
        "Cantidad de ocupados (ESTADO = 1)": [
            (df_2004["ESTADO"] == 1).sum(),
            (df_2024["ESTADO"] == 1).sum(),
            (df_total["ESTADO"] == 1).sum()
        ],
        "Cantidad de desocupados (ESTADO = 2)": [
            (df_2004["ESTADO"] == 2).sum(),
            (df_2024["ESTADO"] == 2).sum(),
            (df_total["ESTADO"] == 2).sum()
        ],
        "Cantidad de variables limpias y homogeneizadas": [
            contar_limpios(df_2004),
            contar_limpios(df_2024),
            contar_limpios(df_total)
        ]
    }

    # Crear tabla estilo PDF
    tabla = pd.DataFrame(datos, index=["Año 2004", "Año 2024", "Total"]).T
    return tabla
#Mostrar la tabla
tabla_final = tabla_resumen_custom(df_2004, df_2024)
print(tabla_final)
```
