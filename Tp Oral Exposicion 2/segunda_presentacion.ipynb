```python
#instalar pandas si no lo tienen. Debes sacar # de #!pip install pandas.
#!pip install pandas
#instalar matplotlib si no lo tienen. Debes sacar # de #!pip install matplotlib.
#!pip install matplotlib
#instalar seaborn si no lo tienen. Debes sacar # de #!pip install seaborn.
#!pip install seaborn
#importar pandas, matplotlib, seaborn y numpy
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
# Leer los archivos
df = pd.read_excel("evyth_microdatos.xlsx")
# Filtrar y separar los de 2024 y 2012
df_2024 = df[df["anio"] == 2024].copy()
df_2012 = df[df["anio"] == 2012].copy()
# Mostrar los datos
display(df_2024)
display(df_2012)
```

```python
# Filtrar solo el trimestre 1
df_2024 = df_2024[df_2024["trimestre"] == 1]
df_2012 = df_2012[df_2012["trimestre"] == 1]
# Mostrar los datos
display(df_2024)
display(df_2012)
```

```python
#Lista de las posibles variables a usar
variables_a_usar= ['id_hogar', 'id_viajes', 'miembro', 'anio', 'trimestre', 'p006', 'nivel_ed', 'region_origen', 'aglomerado_origen', 'region_destino', 'provincia_destino', 'p005', 'cond_act', 'px08_agrup', 'gasto_pc', 'px09']
#p006 es edad
#p005 es sexo
#miembro es Número de miembro en el hogar
#anio es año
#nivel_ed es Máximo nivel educativo alcanzado
#region_origen es Región de residencia del hogar
#aglomerado_origen es Aglomerado de residencia del hogar
#region_destino es Region de destino
#provincia_destino es Provincia de destino
#cond_act es Condición de actividad
#px08_agrup es Tipo de alojamiento (agrupado)
#gasto_pc es gasto total durante el viaje
#px09 es medio de transporte que utilizaron
# Seleccionar las columnas de las variables a usar
df_2024 = df_2024[variables_a_usar]
df_2012 = df_2012[variables_a_usar]
# Mostrar los datos
display(df_2024)
display(df_2012)
```

```python
# Filtrar datos válidos: gasto_pc no nulo y cond_act en 1 (Ocupado) o 2 (Desocupado)
df_plot2012 = df_2012[df_2012["gasto_pc"].notna() & df_2012["cond_act"].isin([1, 2])].copy()
df_plot2024 = df_2024[df_2024["gasto_pc"].notna() & df_2024["cond_act"].isin([1, 2])].copy()

# Mapear etiquetas legibles para graficar
estado_labels = {1: "Ocupado", 2: "Desocupado"}
df_plot2012["estado_labels"] = df_plot2012["cond_act"].map(estado_labels)
df_plot2024["estado_labels"] = df_plot2024["cond_act"].map(estado_labels)
# Kernel por Condición de actividad del 2012
sns.kdeplot(data=df_plot2012[df_plot2012["estado_labels"] == "Ocupado"], x="gasto_pc", label="Ocupado")
sns.kdeplot(data=df_plot2012[df_plot2012["estado_labels"] == "Desocupado"], x="gasto_pc", label="Desocupado")
plt.title("Kernel de gasto total durante el viaje por Condición de actividad del año 2012")
plt.xlabel("Gasto total")
plt.legend()
plt.show()
# Kernel por Condición de actividad del 2024
sns.kdeplot(data=df_plot2024[df_plot2024["estado_labels"] == "Ocupado"], x="gasto_pc", label="Ocupado")
sns.kdeplot(data=df_plot2024[df_plot2024["estado_labels"] == "Desocupado"], x="gasto_pc", label="Desocupado")
plt.title("Kernel de gasto total durante el viaje por Condición de actividad del año 2024")
plt.xlabel("Gasto total")
plt.legend()
plt.show()
```

```python
# Filtrar datos válidos: gasto_pc no nulo y px08_agrup en 1 (Segunda vivienda del hogar) o 2 (Vivienda de familiares y amigos)
df_plot2012 = df_2012[df_2012["gasto_pc"].notna() & df_2012["px08_agrup"].isin([1, 2])].copy()
df_plot2024 = df_2024[df_2024["gasto_pc"].notna() & df_2024["px08_agrup"].isin([1, 2])].copy()

# Mapear etiquetas legibles para graficar
alojamiento_labels = {1: "Segunda vivienda del hogar", 2: "Vivienda de familiares y amigos"}
df_plot2012["alojamiento_labels"] = df_plot2012["px08_agrup"].map(alojamiento_labels)
df_plot2024["alojamiento_labels"] = df_plot2024["px08_agrup"].map(alojamiento_labels)
# Kernel por Tipo de alojamiento (agrupado) del 2012
sns.kdeplot(data=df_plot2012[df_plot2012["alojamiento_labels"] == "Segunda vivienda del hogar"], x="gasto_pc", label="Segunda vivienda del hogar")
sns.kdeplot(data=df_plot2012[df_plot2012["alojamiento_labels"] == "Vivienda de familiares y amigos"], x="gasto_pc", label="Vivienda de familiares y amigos")
plt.title("Kernel de gasto total durante el viaje por Tipo de alojamiento (agrupado) del año 2012")
plt.xlabel("Gasto total")
plt.legend()
plt.show()
# Kernel por Tipo de alojamiento (agrupado) del 2024
sns.kdeplot(data=df_plot2024[df_plot2024["alojamiento_labels"] == "Segunda vivienda del hogar"], x="gasto_pc", label="Segunda vivienda del hogar")
sns.kdeplot(data=df_plot2024[df_plot2024["alojamiento_labels"] == "Vivienda de familiares y amigos"], x="gasto_pc", label="Vivienda de familiares y amigos")
plt.title("Kernel de gasto total durante el viaje por Tipo de alojamiento (agrupado) del año 2024")
plt.xlabel("Gasto total")
plt.legend()
plt.show()
```

```python
# Filtrar datos válidos: gasto_pc no nulo y px09 en 1 (Automóvil o similar (propio)) o 2 (Omnibus)
df_plot2012 = df_2012[df_2012["gasto_pc"].notna() & df_2012["px09"].isin([1, 2])].copy()
df_plot2024 = df_2024[df_2024["gasto_pc"].notna() & df_2024["px09"].isin([1, 2])].copy()

# Mapear etiquetas legibles para graficar
transporte_labels = {1: "Automóvil o similar (propio)", 2: "Omnibus"}
df_plot2012["transporte_labels"] = df_plot2012["px09"].map(transporte_labels)
df_plot2024["transporte_labels"] = df_plot2024["px09"].map(transporte_labels)
# Kernel por medio de transporte usados del año 2012
sns.kdeplot(data=df_plot2012[df_plot2012["transporte_labels"] == "Automóvil o similar (propio)"], x="gasto_pc", label="Automóvil o similar (propio)")
sns.kdeplot(data=df_plot2012[df_plot2012["transporte_labels"] == "Omnibus"], x="gasto_pc", label="Omnibus")
plt.title("Kernel de gasto total durante el viaje por medio de transporte usados del año 2012")
plt.xlabel("Gasto total")
plt.legend()
plt.show()
# Kernel por medio de transporte usados del año 2024
sns.kdeplot(data=df_plot2024[df_plot2024["transporte_labels"] == "Automóvil o similar (propio)"], x="gasto_pc", label="Automóvil o similar (propio)")
sns.kdeplot(data=df_plot2024[df_plot2024["transporte_labels"] == "Omnibus"], x="gasto_pc", label="Omnibus")
plt.title("Kernel de gasto total durante el viaje por medio de transporte usados del año 2024")
plt.xlabel("Gasto total")
plt.legend()
plt.show()
```

```python
# Seleccionar variables para el análisis
vars_pca = ['edad', 'sexo', 'N_miembro_en_hogar', 'nivel_educativo_alcanzado', 'Condición_de_actividad', 'Tipo_de_alojamiento', 'gasto_total', 'medio_de_transporte']
# Filtrar filas completas por año
df_pca_2012 = df_2012[vars_pca].dropna()
df_pca_2024 = df_2024[vars_pca].dropna()

# Matriz de correlación 2012
corr_2012 = df_pca_2012.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_2012, annot=True, cmap="coolwarm", center=0)
plt.title("Matriz de Correlación - Año 2012")
plt.tight_layout()
plt.show()
# Matriz de correlación 2024
corr_2024 = df_pca_2024.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_2024, annot=True, cmap="coolwarm", center=0)
plt.title("Matriz de Correlación - Año 2024")
plt.tight_layout()
plt.show()
```
